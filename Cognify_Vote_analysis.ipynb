```python
# Step 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files

# Step 2: Upload the dataset
uploaded = files.upload()

# Step 3: Load the dataset
df = pd.read_csv("Dataset .csv")

# Step 4: Remove missing data
df_votes = df.dropna(subset=['Votes', 'Aggregate rating'])

# Step 5: Most and Least Voted Restaurants
most_voted = df_votes[['Restaurant Name', 'Votes', 'Aggregate rating']].sort_values(by='Votes', ascending=False).head(5)
least_voted = df_votes[['Restaurant Name', 'Votes', 'Aggregate rating']].sort_values(by='Votes', ascending=True).head(5)

print("üèÜ Most Voted Restaurants:\n", most_voted)
print("\nüï≥Ô∏è Least Voted Restaurants:\n", least_voted)

# Step 6: Correlation between Votes and Rating
correlation = df_votes['Votes'].corr(df_votes['Aggregate rating'])
print(f"\nüìä Correlation between Votes and Ratings: {correlation:.2f}")

# Step 7: Scatter Plot
plt.figure(figsize=(8, 5))
plt.scatter(df_votes['Votes'], df_votes['Aggregate rating'], alpha=0.3, color='purple')
plt.title("Votes vs Aggregate Rating")
plt.xlabel("Votes")
plt.ylabel("Aggregate Rating")
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
