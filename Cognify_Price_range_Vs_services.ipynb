```python
# Step 1: Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 2: Upload dataset
from google.colab import files
uploaded = files.upload()

# Step 3: Load dataset
df = pd.read_csv("Dataset .csv")

# Step 4: Clean relevant columns
df_clean = df.dropna(subset=['Price range', 'Has Online delivery', 'Has Table booking'])

# Step 5: Convert values to uniform format
df_clean['Has Online delivery'] = df_clean['Has Online delivery'].str.strip().str.lower()
df_clean['Has Table booking'] = df_clean['Has Table booking'].str.strip().str.lower()

# Step 6: Calculate percentages
online_delivery_stats = df_clean.groupby('Price range')['Has Online delivery'].value_counts(normalize=True).unstack().fillna(0) * 100
table_booking_stats = df_clean.groupby('Price range')['Has Table booking'].value_counts(normalize=True).unstack().fillna(0) * 100

# Step 7: Plot Online Delivery
plt.figure(figsize=(8, 5))
online_delivery_stats[['yes']].plot(kind='bar', legend=False, color='skyblue')
plt.title("Online Delivery Availability by Price Range")
plt.ylabel("Percentage Offering Delivery")
plt.xlabel("Price Range")
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()

# Step 8: Plot Table Booking
plt.figure(figsize=(8, 5))
table_booking_stats[['yes']].plot(kind='bar', legend=False, color='lightgreen')
plt.title("Table Booking Availability by Price Range")
plt.ylabel("Percentage Offering Table Booking")
plt.xlabel("Price Range")
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
